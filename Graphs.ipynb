## Corpus Le Monde

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from scipy.signal import detrend
import numpy as np
total = pd.read_csv("lemonde1.csv")
neg= pd.read_csv("neg_lemonde.csv")
pos= pd.read_csv("pos_lemonde.csv")

total = total.groupby("annee").sum().reset_index()[["annee","n"]]
total["Année"] = total.annee
total["Negativity"] = neg.n/total.n
total["Positivity"] = pos.n/total.n
total["Emotionality"] = total.Negativity + total.Positivity


fig = px.scatter(total,x="Année",y=["Negativity","Positivity"],template="simple_white",trendline="lowess", trendline_options=dict(frac=0.4))
fig.write_html("lemonde.html")
fig.show()
```

```python
fig = px.scatter(total,x="Année",y="Emotionality", trendline="lowess",template="simple_white", trendline_options=dict(frac=0.2))
fig.write_html("lemonde_emotionality.html")
fig.show()
```

```python
total["Negativity ratio"] = total.Negativity/total.Positivity
fig = px.scatter(total,x="Année",y="Negativity ratio", trendline="lowess",template="simple_white", trendline_options=dict(frac=0.2))
fig.write_html("lemonde_neg_ratio.html")
fig.show()
```

## Corpus Livres Gallica

```python
total = pd.read_csv("base_livres_gallica_monogrammes.csv").sort_values("date").reset_index()
neg= pd.read_csv("neg_livres.csv")
pos= pd.read_csv("pos_livres.csv")


total['Année'] = total.date
total["Negativity"] = neg.n/total.base
total["Positivity"] = pos.n/total.base
total["Emotionality"] = total.Negativity + total.Positivity

total = total.loc[np.logical_and(total.date<1941,total.date>1599),:]

fig = px.scatter(total,x="Année",y=["Negativity","Positivity"],template="simple_white",trendline="lowess", trendline_options=dict(frac=0.3))
fig.write_html("lives.html")
fig.show()
```

```python
px.scatter(total,x="date",y="Emotionality", trendline="lowess",template="simple_white", trendline_options=dict(frac=0.2))
fig.write_html("livres_emotionality.html")
fig.show()
```

```python
total["Negativity ratio"] = total.Negativity/total.Positivity

fig = px.scatter(total,x="date",y="Negativity ratio", trendline="lowess",template="simple_white", trendline_options=dict(frac=0.2))
fig.write_html("livres_neg_ratio.html")
fig.show()
```

## Corpus presse

```python
total = pd.read_csv("base_presse_mois_gallica_monogrammes.csv").sort_values("date").reset_index()
total[["Année","mois"]] = total.date.str.split("/",expand=True)
total = total.groupby("Année").sum().reset_index()
neg= pd.read_csv("neg_presse.csv")
pos= pd.read_csv("pos_presse.csv")
total.Année = total.Année.astype("int")


total["Negativity"] = neg.n/total.base
total["Positivity"] = pos.n/total.base
total["Emotionality"] = total.Negativity + total.Positivity
total = total.loc[np.logical_and(total.Année<1951,total.Année>1788),:]
fig = px.scatter(total,x="Année",y=["Negativity","Positivity"],template="simple_white",trendline="lowess", trendline_options=dict(frac=0.3))
fig.write_html("presse.html")
fig.show()
```

```python
fig = px.scatter(total,x="Année",y="Emotionality", trendline="lowess",template="simple_white", trendline_options=dict(frac=0.3))
fig.write_html("presse_emotionality.html")
fig.show()
```

```python
total["Negativity_ratio"] = total.Negativity/total.Positivity
fig = px.scatter(total,x="Année",y="Negativity_ratio", trendline="lowess",template="simple_white", trendline_options=dict(frac=0.2))
fig.write_html("presse_neg_ratio.html")
fig.show()
```
